```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("random_daily_expenses.csv")

# Display first few rows
print("Initial DataFrame:")
print(df.head()) #first 5
```

```python
# Find the maximum value in the 'Amount' column
max_amount = df["Amount"].max()
print("Maximum Amount:", max_amount)
```

```python
# Get the row corresponding to the max amount
max_row = df[df["Amount"] == max_amount]
print("Row with Maximum Amount:")
print(max_row)
```

```python
# Create an empty list to store top spenders
highest_spenders = []

# Add first spender (the max row)
highest_spenders.append(max_row)

# Remove that spender from the DataFrame
df = df[df["Amount"] != max_amount]

print("Added first highest spender.")
```

```python
# Repeat the process 4 more times
for _ in range(4):
    max_amount = df["Amount"].max()
    max_row = df[df["Amount"] == max_amount]
    highest_spenders.append(max_row)
    df = df[df["Amount"] != max_amount]

print("Top 5 highest spenders have been recorded.")
```

```python
print("Top 5 Highest Spenders:")
for i, spender in enumerate(highest_spenders, start=1):
    print(f"\nSpender {i}:")
    print(spender)
```

```python
# Concatenate all highest spender rows back into one DataFrame
top5_df = pd.concat(highest_spenders)

# Add them back into the original DataFrame
df = pd.concat([df, top5_df], ignore_index=True)

print("Spenders have been re-added to the DataFrame.")
```

```python
category_avg = df.groupby("Category")["Amount"].mean()
print("Average Price per Category:")
print(category_avg)
```

```python
# Map the average price to each row
df["AvgPricePerCategory"] = df["Category"].map(category_avg)
print("DataFrame with AvgPricePerCategory column added:")
print(df.head())
```

```python
# Save to new CSV
df.to_csv("updated_expenses.csv", index=False)

# Save to Excel
df.to_excel("updated_expenses.xlsx", index=False)

print("Files successfully written: updated_expenses.csv and updated_expenses.xlsx")
```
